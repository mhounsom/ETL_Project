### Extract CSVs into DataFrames

```python
players_file = "all_players.csv"
players_df = pd.read_csv(players_file)
players_df
```

```python
salaries_file = "mls_salaries.csv"
salaries_df = pd.read_csv(salaries_file)
salaries_df
```

```python
#combined first and last name columns
salaries_df['Player'] = salaries_df['first_name']+' '+salaries_df['last_name']
salaries_df
```

```python
salaries_df = salaries_df.drop(columns=['club_grouped', 'first_name', 'last_name'])
salaries_df
```

```python
players_df_reg = players_df[players_df["Season"]== 'reg']
players_df_reg
```

```python
players_df_cleaned = players_df_reg[(players_df_reg['Year'] > 2008) & (players_df_reg['Year'] < 2019)]
players_df_cleaned
```
